### 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list

```python
def sum_even():
    l=[]
    n=int(input("Enter how many numbers you want to add in the list: "))
    for i in range(n):
        a=int(input(f"Enter number{i+1}: "))
        l.append(a)
    l1=[item for item in l if item%2==0]
    final=sum(l1)
    return final
sum_even()
```

### 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number

```python
def square(l):
    final=[item**2 for item in l]
    return final
square([1,2,3,4])
```

### 4. Write a Python function that checks if a given number is prime or not from 1 to 200

```python
def prime(n):
    if n>200:
        return "Please enter number between 1 to 200"
    elif n<2:
        return "Not Prime"
    else:
        for i in range(2,int(n**0.5+1)):
            if n%i==0:
                return "Not Prime"
        else:
            return "Prime"
prime(3)
```

### 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms

```python
def fibonacci_iterator(n):
    a,b=0,1
    for _ in range(n):
        yield a
        a,b=b,a+b
for num in fibonacci_iterator(10):
    print(num,end=" ")
```

### 6. Write a generator function in Python that yields the powers of 2 up to a given exponent

```python
def powers_of_two(n):
    for i in range(n+1):
        yield 2**i
for num in powers_of_two(5):
    print(num,end=" ")
```

### 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple

```python
data=[(1,5),(3,1),(2,9),(4,2)]
sorted_data=sorted(data,key=lambda x: x[1])
print(sorted_data)
```

### 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

```python
celsius=[0,20,37,100]
fahernheit=list(map(lambda c: (c*9/5)+32,celsius))
print(fahernheit)
```

### 10. Create a Python program that uses `filter()` to remove all the vowels from a given string

```python
s="Sujal"
vowels='aeiouAEIOU'
no_vowels="".join(filter(lambda ch: ch not in vowels,s))
print(no_vowels)
```
